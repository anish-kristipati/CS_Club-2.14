```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['top50.csv']),encoding = "ISO-8859-1")
```

```python
df=df.head(10)
```

```python
from matplotlib.pyplot import pie, axis, show
```

```python
axis('equal')
pie(df,labels = df.index, rotatelabels = 30,autopct='%.1f%%')
show()
```
